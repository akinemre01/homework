```python

from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python

emredata["AIRLINE"] = label.fit_transform(emredata["AIRLINE"])
emredata["ORIGIN_AIRPORT"] =label.fit_transform(emredata["ORIGIN_AIRPORT"])
emredata["DESTINATION_AIRPORT"] = label.fit_transform(emredata["DESTINATION_AIRPORT"])
```

```python
emredata.columns.values
```

```python
emredata.head(10)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
emredata.isnull().sum()
```

```python

X=emredata.drop(["GECIKME","AIR_TIME","TAXI_OUT","WHEELS_OFF"],axis=1)
y=emredata["GECIKME"]
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1933)
```

```python
logReg = LogisticRegression()
```

```python

logReg.fit(X_train,y_train)
```

```python

y_pred = logReg.predict(X_test)
```

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logReg.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report
```

```python

print(classification_report(y_test,y_pred))
```
